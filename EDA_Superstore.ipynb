## Objective: Perform Exploratory Data Analysis on the dataset, 'SampleSuperstore'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

The dataset contains the following variables: Ship Mode,Segment,Country,Cit,State,Postal Code,Region	Category,Sub-Category,Sales,Quantity,Discount,Profit

```python
df.info()
```

The dataset contains 12 Variables whic are of different types that include object, integer and float

```python
df.describe()      #Summary Stats
```

We can see the the different summary statistics of the variables that are of integer and float type

```python
# Correlation Analysis

corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
```

Here are the following observations:

Negative Correlation- 
i) Sales and Postcode, Sales and Discount, Postal Code and Profit, Profit and Discount

Small Correlations-

ii) Sales and Discount, Sales and Postal code

Medium Correlations-

iii) Sales and Quantity, Sales and Profit

There are no large correlation as no correlation shows values between 0.5 and 1


```python
sns.pairplot(df)
```

The following plot shows all the scatterplots with respect to all the variables. This also helps us understand the correcelation and helps us to visualize it better

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=df,color='r')
plt.show()
```

From this graph, we can observe the various sub categories that are used by each category in the superstore. We can observed that

i) Phones, Accesories, machines and cpiers are used by all the three categorey.

ii) Items like bookcases,chairs,tables and furninshing are not used by any category.

### Comparing the sales value 

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
```

We can observe that Chairs has the highest sales and labels has the lowest sales. Accessories make the most profit out of all sub-categories and tables make the least profit which has made a loss.

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=df)
plt.show()
```

This graph shows the count of quantity of the items sold divided by the region. We can observe that the South region on an average sells very low whereas East sells the highest amongst all regions

```python
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Category')
```
